#### Reflection Questions

On average, do female or male recipients receive larger loans from Kiva?

````python
avg_gender_loan = df.groupby('gender').loan_amount.mean().reset_index()
print(avg_gender_loan)
````

It seems that male recipients recieve larger loans from Kiva. On average, males receive 496.26 USD, while women receive 353.18 USD.

Which country has the *least* disparity in loan amounts awarded by gender?

````python
avg_loan_gendNat = df.groupby(['country','gender']).loan_amount.mean().reset_index()

#cambodia
camb_female = avg_loan_gendNat[(avg_loan_gendNat.gender == 'female') & (avg_loan_gendNat.country == 'Cambodia')]
camb_male = avg_loan_gendNat[(avg_loan_gendNat.gender == 'male') & (avg_loan_gendNat.country == 'Cambodia')]

camb_diff = float(camb_female.loan_amount) - float(camb_male.loan_amount) #positive if female gets more
print('Cambodia Discrepency: ${}\n'.format(camb_diff))

#El Salvador
Sav_female = avg_loan_gendNat[(avg_loan_gendNat.gender == 'female') & (avg_loan_gendNat.country == 'El Salvador')]
Sav_male = avg_loan_gendNat[(avg_loan_gendNat.gender == 'male') & (avg_loan_gendNat.country == 'El Salvador')]

Sav_diff = float(Sav_female.loan_amount) - float(Sav_male.loan_amount)#positive if female gets more
print('El Salvador Discrepency: ${}\n'.format(Sav_diff))

#Kenya
Kenya_female = avg_loan_gendNat[(avg_loan_gendNat.gender == 'female') & (avg_loan_gendNat.country == 'Kenya')]
Kenya_male = avg_loan_gendNat[(avg_loan_gendNat.gender == 'male') & (avg_loan_gendNat.country == 'Kenya')]

Kenya_diff = float(Kenya_female.loan_amount) - float(Kenya_male.loan_amount) #positive if female gets more
print('Kenya Discrepency: ${}\n'.format(Kenya_diff))

#Pakistan
pak_female = avg_loan_gendNat[(avg_loan_gendNat.gender == 'female') & (avg_loan_gendNat.country == 'Pakistan')]
pak_male = avg_loan_gendNat[(avg_loan_gendNat.gender == 'male') & (avg_loan_gendNat.country == 'Pakistan')]

pak_diff = float(pak_female.loan_amount) - float(pak_male.loan_amount)#positive if female gets more
print('Pakistan Discrepency: ${} \n'.format(pak_diff))

#Philippines
fil_female = avg_loan_gendNat[(avg_loan_gendNat.gender == 'female') & (avg_loan_gendNat.country == 'Philippines')]
fil_male = avg_loan_gendNat[(avg_loan_gendNat.gender == 'male') & (avg_loan_gendNat.country == 'Philippines')]

fil_diff = float(fil_female.loan_amount) - float(fil_male.loan_amount)#positive if female gets more
print('Phillipines Discrepency: ${}'.format(fil_diff))

#print(avg_loan_gendNat)
````

## Step 7: Box Plots With Kiva Data

So far you have visualized the average size of loans by country using bar charts; now you are going to make a box plot to compare the distribution of loans by country.

We have set up a figure for you to plot on. Use `sns.boxplot()` to compare the distribution of loan amounts by country for the Kiva dataset.   

`sns.boxplot()` can be passed the same parameters as `sns.barplot()`.

**Optional:** You may set a new color palette if you would like to continue using `sns.set_palette()`.


````python
plt.figure(figsize=(16, 10))

sns.set_context('paper',font_scale = 1.5)
sns.set_palette('Dark2')


ax2 = plt.subplot(111)

sns.boxplot(data=df, x="country", y="loan_amount")

ax2.set_xlabel('Country',fontweight = 'bold')
ax2.set_ylabel('Loan Amount (USD)',fontweight = 'bold')
````

## Step 9: Violin Plots

You can use nearly identical syntax (as you have used for box plots) to create violin plots. Take this line of code from above:

```python
sns.boxplot(data=df, x="activity", y="loan_amount")
```

To visualize the distribution of the exact same data as a violin plot you could pass the same parameters to `sns.violinplot()` instead of `sns.boxplot()`.

Change the code in the cell below so that the data is plotted as a violin plot instead of a barplot.

````python
plt.figure(figsize=(16, 10))

ax_vi_act = plt.subplot(111)

sns.violinplot(data=df, x="activity", y="loan_amount")

ax_vi_act.set_title('Distribution of Loans by Activity')
ax_vi_act.set_xlabel('Activity',fontweight= 'bold')
ax_vi_act.set_ylabel('Loan Amount (USD)',fontweight = 'bold')

````
